```python
data.info()
countrylist = data.country.unique()
```

```python
country_populations = [int(data.population[data.country == country].sum()/data.population[data.country == country].count()) for country in countrylist]
```

```python
plt.rcParams['figure.figsize'] = [20, 30]
plt.barh(countrylist, country_populations)
plt.show()
```

```python
country_populations = np.array(country_populations)
countrylist = np.array(countrylist)
```

```python
sorted_population_perm = country_populations.argsort()
sorted_country = countrylist[sorted_population_perm]
sorted_population = country_populations[sorted_population_perm]
```

```python
sorted_population
```

```python
plt.barh(sorted_country, sorted_population)
plt.show()
```

```python
plt.barh(sorted_country, sorted_population/1000000, color= ["blue", "green"])
plt.xticks(np.arange(min(sorted_population)/1000000, max(sorted_population)/1000000, 100))

plt.show()
```
