```python
import pandas as df
import matplotlib.pyplot as pt
import numpy as np
```

```python
data = df.read_csv('sales.csv')
data.describe()
```

```python
data.axes
```

```python
clean_data = data.drop(columns=data.columns[0])#drop first column which is index not needed
clean_data.describe()
```

```python
fig = pt.figure()
tuesday = fig.add_subplot(2,2,1)
pt.plot(np.arange(1,53), data['tuesday'][0:52] )#slice only first 52 weeks which ,makes onne year
```

```python
fig = pt.figure()
tuesday = fig.add_subplot(2,2,1)
pt.plot(np.arange(1,53), data['tuesday'][0:52] )
```

```python
pt.plot(np.arange(1,53), data['monday'][0:52], color = 'red', label = 'monday')
pt.plot(np.arange(1,53), data['tuesday'][0:52], color = 'black', label = 'Tuesday')
pt.plot(np.arange(1,53), data['wednesday'][0:52], color = 'yellow', label = 'wednesday')
pt.plot(np.arange(1,53), data['thursday'][0:52], color = 'orange', label = 'thursday')
pt.plot(np.arange(1,53), data['friday'][0:52], color = 'maroon', label = 'friday')
pt.plot(np.arange(1,53), data['saturday'][0:52], color = 'green', label = 'saturday')
pt.plot(np.arange(1,53), data['sunday'][0:52], color = 'gray', label = 'sunday')
pt.rcParams['figure.figsize'] = [1,2]
pt.xlabel('Week')
pt.ylabel('Sales in Naira (x10)')
pt.title('Weekly Sales Chart Year One')
pt.xlim(0, 60)
pt.legend()
pt.show()
pt.savefig('p.pdf')

```

```python
# data['tuesday'][0:52]
fig = pt.figure()
pt.rcParams['figure.figsize'] = [10, 8]
tuesday = fig.add_subplot(4,1,1)
tuesday.plot(np.arange(1,53), data['tuesday'][0:52], color = 'red', label = 'Tuesday' )
wednesday = fig.add_subplot(4,1,2)
wednesday.plot(np.arange(1,53), data['wednesday'][0:52], color = 'maroon', label = 'wednesday' )
thursday = fig.add_subplot(4,1,3)
thursday.plot(np.arange(1,53), data['thursday'][0:52], label = 'thursday' )
thursday.set_xlabel('Sales')
thursday.set_ylabel('Week')
friday = fig.add_subplot(4,1,4)
friday.plot(np.arange(1,53), data['friday'][0:52], color = 'green', label = 'Friday' )
friday.set_xlabel('Sales')
friday.set_ylabel('Week')

fig.legend()
pt.show()
pt.savefig('p.pdf')
```

```python
fig = pt.figure()
pt.rcParams['figure.figsize'] = [10, 12]
tuesday = fig.add_subplot(4,1,1)
tuesday.plot(np.arange(1,53), data['tuesday'][0:52], color = 'red', label = 'Tuesday' )
wednesday = fig.add_subplot(4,1,1)
wednesday.plot(np.arange(1,53), data['wednesday'][0:52], color = 'maroon', label = 'wednesday' )
thursday = fig.add_subplot(4,1,1)
thursday.plot(np.arange(1,53), data['thursday'][0:52], label = 'thursday' )
thursday.set_xlabel('Sales')
thursday.set_ylabel('Week')
saturday = fig.add_subplot(4,1,1)
saturday.plot(np.arange(1,53), data['saturday'][0:52], color = 'yellow', label = 'saturday' )
saturday.set_xlabel('Sales')
saturday.set_ylabel('Week')

fig.legend()
pt.show()
pt.savefig('p.pdf')
```

```python
monsal = data['monday'].sum()
tuesal = data['tuesday'].sum()
wedsal = data['wednesday'].sum()
thursal = data['thursday'].sum()
frisal = data['friday'].sum()
satsal = data['saturday'].sum()
sunsal = data['sunday'].sum()
```

```python
pt.bar(['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'saturday', 'Sunday'],[monsal,tuesal,wedsal, thursal, frisal, satsal, sunsal])
pt.show()
```
